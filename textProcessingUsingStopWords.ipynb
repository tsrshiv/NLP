```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4') # Recommended for better compatibility

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("going", pos="v")) # 'v' tells it 'going' is a verb
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
# English stopwords ki list dekhne ke liye
print(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
stop_words= set(stopwords.words('english'))
text="Hello my name is shivendra and i am from jalalabad shahjahanpur."
words=word_tokenize(text.lower())
filtered_sentence= [w for w in words if w not in stop_words]
print(filtered_sentence)
```
