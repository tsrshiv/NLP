```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import pandas as pd
```

```python
docs=["Cricked is played with a bat. Cricket is a gebtkeman game",
      "Football is a popular sport"
      ]
```

```python
vectorizer= TfidfVectorizer()
```

```python
tfidf_matrix=vectorizer.fit_transform(docs)
```

```python
feature_names=vectorizer.get_feature_names_out()
```

```python
df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names, index=["Doc 1", "Doc 2"])

print(df)
```
