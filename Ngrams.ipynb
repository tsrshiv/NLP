```python
import nltk
from nltk.util import ngrams
```

```python
text="mujhe coding bahaut psnd h"
```

```python
nltk.download('punkt_tab')
```

```python
tokens = nltk.word_tokenize(text)
```

```python
bigrams= list(ngrams(tokens,2))
```

```python
trigrams= list(ngrams(tokens,3))
```

```python
print(tokens)
```

```python
print(bigrams )
```
