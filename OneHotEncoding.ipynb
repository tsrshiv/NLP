```python
import pandas as pd

# Ek chota sa dataset banate hain
data = {'Fruit': ['Apple', 'Mango', 'Orange', 'Apple']}
df = pd.DataFrame(data)

# One-Hot Encoding apply karte hain
# drop_first=True karne se "Dummy Variable Trap" se bach sakte hain
df_encoded = pd.get_dummies(df, columns=['Fruit'], prefix='Fruit')

print(df_encoded)
```
